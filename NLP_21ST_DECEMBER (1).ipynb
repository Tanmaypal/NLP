```
import pandas as pd
import regex as re
```

```
dataframe=pd.read_json("http://snap.stanford.edu/data/amazon/productGraph/categoryFiles/reviews_Office_Products_5.json.gz",compression='infer',lines=True)
```

```
dataframe.head()
```

```
##Cleaning of the review coloum
def cleaning(text):
  return re.sub('[^a-zA-Z]'," ",text)
```

```
dataframe['reviewText2'] =dataframe['reviewText'].apply(lambda text: cleaning(text))
dataframe['reviewText2']
```

```
results = set()
dataframe['reviewText2'].str.lower().str.split().apply(results.update)
print(len(results))
```

```
arr=list(results)
arr1=arr[1:301]
len(arr1)
```

```
from itertools import combinations
def rSubset(arr, r): 
  
    # return list of all subsets of length r 
    # to deal with duplicate subsets use  
    # set(list(combinations(arr, r))) 
    return list(combinations(arr, r))
```
