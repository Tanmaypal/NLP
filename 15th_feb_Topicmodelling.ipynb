```
list1=[]
for i in range(0,len(df2)):
  review=re.sub('[^a-zA-Z]',' ',df2['review_combined'][i])
  review=review.lower()
  review=word_tokenize(review)
   #review=review.split()
  pos_tag=nltk.pos_tag(review)
  list1.append(pos_tag)
```

```
list1[0]
```

```
def get_wordnet_pos(treebank_tag):
    if treebank_tag.startswith('J'):
        return wordnet.ADJ
    elif treebank_tag.startswith('V'):
        return wordnet.VERB
    elif treebank_tag.startswith('N'):
        return wordnet.NOUN
    elif treebank_tag.startswith('R'):
        return wordnet.ADV
    else:
        return 'n'
  
```

```
list1
```

```
lemma=[]

for word in list1:
  #print(word)
  l=[]
  for p in word:
    #print(p)
    m=Lemmitizer.lemmatize(p[0], get_wordnet_pos(p[1]))
    #print(p, m)
    l.append(m)

  l = " ".join(l)
  lemma.append(l)
```

```
lemma
```

```
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer()
f=tfidf.fit_transform(lemma)
features=pd.DataFrame(f.toarray(),columns=tfidf.get_feature_names())
```

```
W=model.components_
```

```
import numpy as np
```

```
def display_topics(H,W,feature_names, documents, no_top_words, no_top_documents):
    for topic_idx, topic in enumerate(H):
        print("Topic %d:" % (topic_idx))
        print(" ".join([feature_names[i]
                        for i in topic.argsort()[:-no_top_words - 1:-1]]))
        top_doc_indices = np.argsort( W[:,topic_idx] )[::-1][0:no_top_documents]
        for doc_index in top_doc_indices:
            print(documents[doc_index])
```

```
display_topics(W,H,tfidf.get_feature_names(),df1['review_combined'], 4,3)
```
